```python
# example of image augmentation with multiple transformations
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
#Prepare the image
# a PIL image is returned
cat_img = load_img('images/cat.jpg')

# get a Numpy array with shape (3, 200, 200)
# (n_channels, y_pixels, x_pixels)
cat_img_array = img_to_array(cat_img) 

# Keras requires a 4D array with all images in it (n_images, n_channels, y_pixels, x_pixels)
# returns a Numpy array with shape (1, 3, 200, 200), 1 since we have only 1 image
cat_img_array = cat_img_array.reshape((1,) + cat_img_array.shape)  
```

```python
# example of image augmentation with multiple transformations
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
# construct an instance of the ImageDataGenerator class
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        brightness_range=[0.2,2.0],
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
import matplotlib.pyplot as plt

i = 0
for batch in datagen.flow(cat_img_array, batch_size=1):
    
    # TO VIEW IMAGE INLINE
    # convert to unsigned integers for plotting
    image = batch[0].astype('uint8')
    # plot raw pixel data
    plt.imshow(image)
    # show the figure
    plt.show()
    
    i += 1
    if i > 20:
        break  # break loop at the count of 20
```

```python
# example of image augmentation with multiple transformations
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
# construct an instance of the ImageDataGenerator class
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        brightness_range=[0.2,2.0],
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
import matplotlib.pyplot as plt
import numpy as np

i = 0
for batch in datagen.flow_from_directory(directory='images', batch_size=1):
    
    # TO VIEW IMAGE INLINE
    # convert to unsigned integers for plotting
    image = batch[0].astype('uint8')
    
     # changing size from (1, 200, 200, 3) to (200, 200, 3) for plotting the image
    image=np.delete(image,1)

    # plot raw pixel data
    plt.imshow(image)
    # show the figure
    plt.show()
    
    i += 1
    if i > 20:
        break  # break loop at the count of 20
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt
import numpy as np
```

```python
#import all VGG specific libraries
from tensorflow.keras.applications.inception_v3 import InceptionV3, preprocess_input, decode_predictions
```

```python
#load the InceptionV3 model
model = InceptionV3(weights='imagenet', include_top=True)
model.summary()
```

```python
#load an image for prediction
img = load_img('images/ship.jpg', target_size=(299,299))
plt.imshow(img)
plt.show()
```

```python
#covert img to array and add dimension
img = img_to_array(img)
img = img.reshape(1,299,299,3)
```

```python
#getting the prediction
result = model.predict(img)
```

```python
#decoding proablity values into string labels
class_label = decode_predictions(result)
```

```python
print(class_label)
```

```python
class_label_single = class_label[0][0]
print("the prediction is:")
print(class_label_single)
```

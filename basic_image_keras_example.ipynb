```python
#importing load_image
from tensorflow.keras.preprocessing.image import load_img, img_to_array, array_to_img, save_img
```

```python
#load image from folder and convert to PIL object
cat_img = load_img('images/cat.jpg')

#print properties of image
print(cat_img.size)
print(type(cat_img))
print(cat_img.mode)
print(cat_img.format)
```

```python
#show image in viewer
cat_img.show()
```

```python
#show image inline
import matplotlib.pyplot as plt

plt.imshow(cat_img)
plt.show()
```

```python
#converting the PIL image to numpy array
cat_img = img_to_array(cat_img)
```

```python
#printing the properties of numpy array
print(type(cat_img))
print(cat_img.shape)
print(cat_img)
```

```python
#convert numpy array to pil image
cat_img = array_to_img(cat_img)
```

```python
#importing load_image
from tensorflow.keras.preprocessing.image import load_img, img_to_array

#load image from folder and convert to PIL object
cat_img = load_img('images/cat.jpg')
#convert numpy array to pil image
cat_img = img_to_array(cat_img)
print(cat_img.shape)
```

```python
#load image from folder and convert to PIL object
cat_img = load_img('images/cat.jpg', color_mode='grayscale')
#convert numpy array to pil image
cat_img = img_to_array(cat_img)
print(cat_img.shape)
```

```python
import numpy as np
#change channel order of array
cat_img = np.rollaxis(cat_img, 2, 0)
print(cat_img.shape)
```

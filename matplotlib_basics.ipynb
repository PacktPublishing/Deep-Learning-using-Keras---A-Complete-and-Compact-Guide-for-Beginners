```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#line plot
myarray = np.array([1,2,3,2,10])
plt.plot(myarray)
plt.xlabel("x axis values")
plt.ylabel("y axis values")
plt.show()
```

```python
#bar plot
myarray = np.array([10,10,20,30,30,30,40,50])
plt.hist(myarray)
plt.xlabel("x axis values")
plt.ylabel("y axis values")
plt.show()
```

```python
#scatter plot
myarray1 = np.array([10,20,30,40,50,30,40,50])
myarray2 = np.array([60,70,80,90,30,30,40,50])
plt.scatter(myarray1,myarray2)
plt.xlabel("x axis values")
plt.ylabel("y axis values")
plt.show()
```
